### Exercice n°1

Dans un processeur 32 bits, la taille des registres est de 32 bits. Combien d'adresses mémoire peuvent être référencées dans un registre ? Et pour un processeur 64 bits? En déduire pourquoi les ordinateurs de bureau actuellement en vente ont tous des processeurs 64 bits.


**Correction**:

Dans un processeur 32 bits, les registres peuvent référencer $2^{32}$ adresses différentes. C'est-à-dire que la taille de la mémoire ne pourra pas excéder $2^{32} \sim 4 Go$. Un processeur 64 bits peut référencer $2^{64}$ adresses mémoire différentes. La mémoire des ordinateurs actuels peut aller jusqu'à 16 Go. Un processeur 32 bits sera donc insuffisant.

### Exercice n°6

On considère le programme suivant:

```python
a = int(input()) # entier lu stocké dans le registre R0
b = int(input()) # entier lu stocké dans le registre R1
if a > b:
	m = a
else:
	m = b
# le maximum m de a et b est stocké dans le registre R2
# et en mémoire centrale à l'adresse 20
print(m)
```

Traduire ce programme en assembleur puis le tester dans le simulateur.


**Correction:**

```assembly
      //Lecture
      INP R0, 2
      INP R1, 2
      //a > b
      CMP R0, R1
      //Branchement conditionnel sur l'étiquette suite si a > b
      BGT suite
      MOV R2, R1
      //Branchement inconditionnel sur l'étiquette fin
      B fin
      //étiquette suite
suite:
      MOV R2, R0
      //étiquette fin
fin:
      //affichage du registre R1 dans Output
      OUT R2, 4
      HALT
```
